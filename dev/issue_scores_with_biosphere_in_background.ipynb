### 1) biosphere flows in 'deep' background and in the foreground system, but not at the temporal market process

```python
from tests.databases import *
db_abcd_CO2_foreground_and_in_deep_background()


from datetime import datetime
database_date_dict = {
    "background_2008": datetime.strptime("2008", "%Y"),
    "background_2024": datetime.strptime("2024", "%Y"),
    "foreground": "dynamic",  # flag databases that should be temporally distributed with "dynamic"
}

SKIPPABLE = [
    node.id for node in bd.Database('background_2008')] + [
    node.id for node in bd.Database('background_2024')
]

def filter_function(database_id: int) -> bool:
    return database_id in SKIPPABLE

demand = {('foreground', 'A'): 1}
method = ('GWP', 'example')

from medusa import MedusaLCA
lca2 = MedusaLCA(demand, method, filter_function, database_date_dict, max_calc=5000)

import warnings
warnings.filterwarnings("ignore")
lca2.build_timeline()

lca2.build_datapackage()

lca2.lci()
lca2.lcia()

print('Old static LCA Score:', lca2.static_lca.score)
print('New MEDUSA LCA Score:', lca2.score)
```

```python
lca2.create_labelled_technosphere_dataframe()
```

#### -> in this case, everything seems to work
